```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the uploaded Excel file
file_path = 'Surge_Afternoon_High_Magnitude(6d).xlsx'
data = pd.ExcelFile(file_path)

df = data.parse('Sheet1')

# Extract necessary columns
df['Hour'] = df['Hour'].astype(float)
df_grouped = df.groupby(['Date', 'Hour'])['USEP (c/kWh)'].mean().unstack(level=0)

# Calculate the average price for each hour across all days
average_price = df_grouped.mean(axis=1)

# Plot the price for each day
plt.figure(figsize=(14, 7))
for day in df_grouped.columns:
    plt.plot(df_grouped.index, df_grouped[day], color='gray', alpha=0.5)

# Plot the average price line
plt.plot(df_grouped.index, average_price, color='blue', label='Average Price', linewidth=2)

# Formatting the plot
plt.title('Electricity Price vs. 24-Hour Day', fontsize=16)
plt.xlabel('Hour of Day', fontsize=14)
plt.ylabel('Electricity Price (c/kWh)', fontsize=14)
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```

```python
file_path_new = 'Super_flat_days_under_Q3 (29.49 per cent).xlsx'
data_new = pd.ExcelFile(file_path_new)

df_new = data_new.parse('Sheet1')

# Convert Datetime column to datetime format and extract relevant information
df_new['Datetime'] = pd.to_datetime(df_new['Datetime'])
df_new['Hour'] = df_new['Datetime'].dt.hour + df_new['Datetime'].dt.minute / 60
df_new['Date'] = df_new['Datetime'].dt.date

# Pivot the data to have hours as rows and dates as columns
df_grouped_new = df_new.pivot(index='Hour', columns='Date', values='USEP (c/kWh)')

# Calculate the average price for each hour across all days
average_price_new = df_grouped_new.mean(axis=1)

# Plot the price for each day
plt.figure(figsize=(14, 7))
for day in df_grouped_new.columns:
    plt.plot(df_grouped_new.index, df_grouped_new[day], color='gray', alpha=0.5)

# Plot the average price line
plt.plot(df_grouped_new.index, average_price_new, color='blue', label='Average Price', linewidth=2)

# Formatting the plot
plt.title('Electricity Price vs. 24-Hour Day', fontsize=16)
plt.xlabel('Hour of Day', fontsize=14)
plt.ylabel('Electricity Price (c/kWh)', fontsize=14)
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
